```python
def country_to_continent(country):
    for continent, countries in continents_data.items():
        if country in (countries):
            return continent
    return None

```

```python
dataFrames={}
for i in range(len(file_list)):
    year=file_list[i][4:8]
    file_path=f"/home/jovyan/workspace/DATASET/WHR_{year}.csv"
    dataFrames[f'd{i}']=pd.read_csv(file_path)
    dataFrames[f'd{i}']['Year']=year
    dataFrames[f'd{i}']['Continent']=dataFrames[f'd{i}']['country'].apply(country_to_continent)
    
    
```

```python
from geopy.geocoders import Photon
geolocator = Photon(user_agent="measurements")
```

```python
#geolocator = Nominatim(user_agent="myGeocoder") -------- nie dziala
def geo_location(country: str):
    try:
        location = geolocator.geocode(country)
        if location is not None:
            return pd.Series([location.latitude, location.longitude])
        else:
            return pd.Series([None, None])
    except Exception as e:
        return pd.Series([None, None])
    

    
df[['lat', 'long']] = df['country'].apply(geo_location)

```

```python
#interactive pipeline
idf=df.interactive()
```

```python
#panel
year_slider=pn.widgets.IntSlider(name='Year Slider',start=2015,end=2023,step=1,value=2021)
year_slider
```

```python
yaxis=pn.widgets.RadioButtonGroup(
    name='Y axis',
    options=['happiness_score','social_support', 'generosity'],
    button_type='success')

    
    
```

```python
continents = ['Asia','Europe','North America','South America','Africa','Oceania']
happiness_pipeline=(
    idf[(idf.Year<=year_slider)]
    .groupby(['Continent','Year'])[yaxis].mean()
    .to_frame()
    .reset_index()
    .sort_values(by='Year')
    .reset_index(drop=True)
)
```
